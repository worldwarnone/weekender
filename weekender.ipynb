```python
import os
import pandas as pd
import plotly.express as px
import datetime
```

```python
btc_eur = pd.read_csv('/kaggle/input/392-crypto-currency-pairs-at-minute-resolution/btceur.csv')
btc_eur_close = btc_eur.loc[:, ['time', 'close']]
btc_eur_close.rename(columns={'close': 'btc_eur'}, inplace=True)
btc_eur_close.head()
```

```python
btc_usd = pd.read_csv('/kaggle/input/392-crypto-currency-pairs-at-minute-resolution/btcusd.csv')
btc_usd_close = btc_usd.loc[:, ['time', 'close']]
btc_usd_close.rename(columns={'close': 'btc_usd'}, inplace=True)
btc_usd_close.head()
```

```python
btc_eur_usd = pd.merge(btc_eur_close, btc_usd_close, on='time', how='inner').sort_values(by='time')
btc_eur_usd['eur_usd'] = btc_eur_usd.btc_usd / btc_eur_usd.btc_eur
btc_eur_usd['time'] = pd.to_datetime(btc_eur_usd.time * 1000000)
btc_eur_usd['day_of_week'] = btc_eur_usd.time.dt.dayofweek 
btc_eur_usd['time_of_day'] = btc_eur_usd.time.dt.time
btc_eur_usd['is_weekend'] = 0
btc_eur_usd.is_weekend[(btc_eur_usd.day_of_week == 6) & (btc_eur_usd.time_of_day < datetime.time(20))] = 1
btc_eur_usd.is_weekend[btc_eur_usd.day_of_week == 5] = 1
btc_eur_usd['last_price'] = btc_eur_usd.eur_usd
btc_eur_usd.last_price[btc_eur_usd.is_weekend == 1] = None
btc_eur_usd.ffill(inplace=True)
btc_eur_usd.tail(50)
```

```python
dec_22_data = (
    btc_eur_usd
    .loc[:, ['time', 'eur_usd', 'is_weekend']]
    .set_index('time')
    .loc['10-2022']
    .reset_index()
)

dec_22_data.head()
```

```python
fig = px.scatter(dec_22_data, x='time', y='eur_usd', color='is_weekend')
fig.show()
```

```python
test_data = btc_eur_usd.set_index('time')['2018':]
test_data = test_data[test_data.is_weekend == 1].sort_values('time')

test_data.head()
```

```python
position = []
trade_counter = 0

pos = 0
for i in range(len(test_data)):
    entry = test_data.iloc[i, :]
    if pos == 0:
        if entry.eur_usd - entry.last_price > .003:
            pos = -1
            trade_counter += 1
        if entry.eur_usd - entry.last_price < -.003:
            pos = 1
            trade_counter += 1
    elif pos == 1:
        if entry.eur_usd - entry.last_price > -.0005:
            pos = 0
    elif pos == -1:
        if entry.eur_usd - entry.last_price < .0005:
            pos = 0
    position.append(pos)

test_data['position'] = position
```

```python
trade_data = test_data.loc[:, ['eur_usd', 'position']]
trade_data.eur_usd = trade_data.eur_usd.pct_change()
trade_data.position = trade_data.position.shift(1)
trade_data.dropna(inplace=True)

trade_data.head()
```

```python
trade_data['min_pnl'] = trade_data.eur_usd * trade_data.position
trade_data['cumulative_pnl'] = trade_data.min_pnl.expanding().apply(lambda x: np.prod(1 + x) - 1)

trade_data.head()
```

```python
fig = px.line(trade_data.reset_index(), x='time', y='cumulative_pnl')
fig.show()
```
